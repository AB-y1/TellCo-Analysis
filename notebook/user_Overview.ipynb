```python

# 1. Get manufacturer counts
manufacturer_counts = telecom_df['Handset Manufacturer'].value_counts().head(3)  # Top 3 manufacturers

# 2. Iterate through top manufacturers
for manufacturer in manufacturer_counts.index:
  # Filter data for current manufacturer
  filtered_df = telecom_df[telecom_df['Handset Manufacturer'] == manufacturer]
  
  # Get top 5 handsets (by count) for this manufacturer
  top_5_handsets = filtered_df['Handset Type'].value_counts().head(5)
  
  # Print results
  print(f"Top 5 Handsets for {manufacturer}:")
  print(top_5_handsets)
  print(manufacturer_counts.index)
  print("\n")  # Add a newline for better formatting
```

```python

# 1. Get manufacturer counts
manufacturer_counts = telecom_df['Handset Manufacturer'].value_counts().head(3)

# 2. Iterate and Plot
for manufacturer in manufacturer_counts.index:
  # Filter data
  filtered_df = telecom_df[telecom_df['Handset Manufacturer'] == manufacturer]
  
  # Get top 5 handsets
  top_5_handsets = filtered_df['Handset Type'].value_counts().head(5)
  
  # Create a new figure (using object-oriented API)
  fig, ax = plt.subplots()  # Create a figure and an axes object
  top_5_handsets.plot(kind='bar', color='skyblue', ax=ax)  # Plot on the axes object
  plt.title(f"Top 5 Handsets for {manufacturer}")
  plt.xlabel("Handset Type")
  plt.ylabel("Count")
  plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
  plt.tight_layout()
  
  plt.show()
```
